# Ch 10 Eigenvalue Adjustment

```python
# numerical and scientific computing libraries
import numpy as np
import scipy as sp

# plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# for pretty printing
np.set_printoptions(4, linewidth=100, suppress=True)
```

```python
# create a random stochastic matrix with size n.
def random_stochastic_matrix(n):
    A = np.random.rand(n, n)
    A = A / np.sum(A, axis=1)[:, None]
    return A
```

```python
n = 10
alpha = 0.01

# create a random stochastic matrix
A = random_stochastic_matrix(n)
B = A + alpha * np.ones((n,1)) @ np.ones((1,n))
C = A - alpha * np.ones((n,1)) @ np.ones((1,n))

# extract eigenvalues and eigenvectors.
eigenvalues_A, eigenvectors_A = sp.linalg.eig(A)
eigenvalues_B, eigenvectors_B = sp.linalg.eig(B)
eigenvalues_C, eigenvectors_C = sp.linalg.eig(C)

# normalize the eigenvectors
eigenvectors_A = np.sign(eigenvectors_A[0,0])* eigenvectors_A
eigenvectors_B = np.sign(eigenvectors_B[0,0])* eigenvectors_B
eigenvectors_C = np.sign(eigenvectors_C[0,0])* eigenvectors_C

# compare the eigenvalues
print("Eigenvalues of A:")
print(eigenvalues_A)
print("Eigenvalues of B:")
print(eigenvalues_B)
print("Eigenvalues of C:")
print(eigenvalues_C)
# compare the first eigenvectors
print("The first eigenvectors of A:")
print(eigenvectors_A[:, 0])
print("The first eigenvectors of B:")
print(eigenvectors_B[:, 0])
print("The first eigenvectors of C:")
print(eigenvectors_C[:, 0])
# plot the eigenvalues
plt.figure(figsize=(10, 5))
plt.scatter(eigenvalues_A.real, eigenvalues_A.imag, color='red', label='Eigenvalues of A')
plt.scatter(eigenvalues_B.real, eigenvalues_B.imag, color='green', label='Eigenvalues of B')
plt.scatter(eigenvalues_C.real, eigenvalues_C.imag, color='blue', label='Eigenvalues of C')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Eigenvalues of Stochastic Matrices')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid()
plt.legend()
```
